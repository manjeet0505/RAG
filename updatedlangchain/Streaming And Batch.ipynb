### Streaming

Streaming means sending output incrementally as it's generated, rather than waiting for the entire response to complete before showing anything. Instead of: generate everything → then send, it's: generate a token → send it → generate next token → send it...

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("Why do parrots talk?")
response.content
```

```python
for chunk in model.stream("Why Do parrots Colorful Feathers?"):
 print(chunk.text,end="|",flush=True)
```

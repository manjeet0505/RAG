### Streaming

Streaming means sending output incrementally as it's generated, rather than waiting for the entire response to complete before showing anything. Instead of: generate everything → then send, it's: generate a token → send it → generate next token → send it...

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google-genai:gemini-2.5-flash-lite")
response = model.invoke("Why do parrots talk?")
response.content
```

```python
for chunk in model.stream("Why Do parrots Colorful Feathers?"):
 print(chunk.text,end="|",flush=True)
```

### Batch

Batching means grouping multiple requests together and processing them at once, rather than sending them one by one. It's essentially the opposite philosophy to streaming.

```python
responses = model.batch(
    "Why Do parrots Colorful Feathers?"
    "What is the main aim of napoleon?"
)
for response in responses:
    print(response)
```
